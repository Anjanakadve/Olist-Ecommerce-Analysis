```python
def contains_number(s):
    return any(i.isdigit() for i in  str(s))
```

```python
geolocation[geolocation["geolocation_city"].apply(contains_number)]
```

```python
import re

def remove_numbers(s):
    return re.sub(r'\d+','',str(s))
```

```python
geolocation["geolocation_city"] = geolocation["geolocation_city"].apply(remove_numbers)
```

```python
geolocation[geolocation["geolocation_city"].apply(contains_number)]
```

```python
 geolocation['geolocation_city']=geolocation["geolocation_city"].str.replace("'","",regex=False)
```

```python
geolocation.duplicated().sum()
```

```python
geolocation[geolocation["geolocation_city"].str.contains("'", na=False)]
```

```python
geolocation[geolocation["geolocation_city"].str.startswith("*")]
```

```python
geolocation["geolocation_city"]= geolocation["geolocation_city"].str.replace("*","", regex=False)
```

```python
geolocation[geolocation["geolocation_city"].str.startswith("*")]
```

```python
def keep_only_text(x):
    if pd.isna(x):
        return x
    # remove everything except alphabets and spaces
    return re.sub(r"[^A-Za-z\s]", "", str(x)).strip()
```

```python
geolocation["geolocation_city"]= geolocation["geolocation_city"].apply(keep_only_text)
```

```python
geolocation["geolocation_city"]= geolocation["geolocation_city"].str.replace("Sao Joao Do Pau D Alho","Sao Joao Do Pau DAlho")
```

```python
geolocation.drop_duplicates(inplace=True)
```
